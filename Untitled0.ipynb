```python
import pandas as pd  # 拿出一个叫 Pandas 的超级计算器
import seaborn as sns # 拿出一个叫 Seaborn 的画图工具
```

```python
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
# 意思：电脑，帮我打开那个 Excel 文件，以后我就叫它 "df" (Dataframe 的缩写)。
```

```python
df['Attrition_Num'] = df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
# 意思：Excel 里写的是 "Yes" 和 "No"，电脑不好算平均数。
# 我让电脑把 "Yes" 变成数字 1，把 "No" 变成数字 0。
# 这样我算平均值得出 0.3，就代表 30% 的离职率。
```

```python
overtime_attrition = df.groupby('OverTime')['Attrition_Num'].mean()
# 意思：
# groupby('Overtime'): 把人按“加班”和“不加班”分成两堆。
# ['Attrition_Num'].mean(): 算出每一堆里，离职（就是那个1）的平均比例是多少。
```

```python
overtime_attrition
```

```python
overtime_df = overtime_attrition.reset_index()
```

```python
import matplotlib.pyplot as plt

sns.barplot(
    data=overtime_df,
    x='OverTime',
    y='Attrition_Num'
)

plt.title('Attrition Rate by Overtime Status')
plt.ylabel('Attrition Rate')
plt.xlabel('Overtime')
plt.show()
```

```python
df.groupby('Department')['Attrition_Num'].mean()
```

```python
dept_attrition = df.groupby('Department')['Attrition_Num'].mean()
```

```python
department_df=dept_attrition.reset_index()
```

```python
dept_attrition
```

```python
sns.barplot(
    data=department_df,
    x=dept_attrition.index,
    y=dept_attrition.values,
)
plt.title('Attrition Rate by Department')
plt.ylabel('Attrition Rate')
plt.xlabel('Department')

plt.show()
```

```python
bins = [0, 3000, 6000, df['MonthlyIncome'].max()]
labels = ['Low Income', 'Mid Income', 'High Income']

df['Income_Band'] = pd.cut(
    df['MonthlyIncome'],
    bins=bins,
    labels=labels
)
```

```python
df.groupby('Income_Band')['Attrition_Num'].mean()
```

```python
income_attrition=(
    df.groupby('Income_Band', observed=True)['Attrition_Num'].mean()
)
```

```python
income_df=income_attrition.reset_index()
```

```python
income_attrition
```

```python
sns.barplot(
    data=income_df,
    x='Income_Band',
    y='Attrition_Num'
)
plt.title('Attrition Rate by Income Band')
plt.ylabel('Attrition Rate')
plt.xlabel('Income Band')

plt.show()
```
